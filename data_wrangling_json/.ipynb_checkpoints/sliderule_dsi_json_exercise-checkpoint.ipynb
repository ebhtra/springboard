## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Load the data into a DataFrame
json_df = pd.read_json('data/world_bank_projects.json')
```

## 1. Find the 10 countries with most projects

```python
json_df.countryshortname.value_counts()[:8].append(
    json_df.countryshortname.value_counts()[9:11]) 
# 9th item in value_counts(), Africa, is continent, so is not included in this answer
```

```python
# Visualize results, to learn some plotting techniques

import matplotlib.pyplot as plt
%matplotlib inline

json_df.countryshortname.value_counts()[::-1].plot(kind='barh', figsize=(6,20), xticks=range(0,20,2), grid=True)
plt.text(10,100,'Projects per Country', bbox=dict(facecolor='lightblue'), fontsize=14)
plt.tick_params(top=True, bottom=True, labeltop=True)
plt.show()
```

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
with open('data/world_bank_projects.json') as f:
    j = json.load(f)

j_norm = json_normalize(j, 'mjtheme_namecode')
j_norm.code.value_counts()[:10]
```

### Question 2 probably wants the names of the themes, not the codes.

```python
# So find if the codes all map to a unique name, after dropping blanks and dupes.
codemap = j_norm[j_norm['name'] != ''].drop_duplicates()
codemap
```

```python
# Apparently all 11 codes map to just one name each, so that makes it easy
codemap = codemap.set_index('code').to_dict()['name']
codemap
```

### Use a Counter to answer Q.2

```python
from collections import Counter
# Use the codemap from above cell to fill in the blanks
codelist = list(map(lambda code: codemap[code], j_norm.code))
Counter(codelist).most_common()[:10]
```

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# This was basically done to answer Q.2
# Just replace the partially empty 'name' column with 'codelist' from Q.2
j_norm['name'] = codelist
j_norm
```

### But maybe Q.3 is looking for a DataFrame with all the original data, plus filled in names:

```python
# apply this func to each row of the mjtheme_namecode column
def fill_names(mjtheme_cell):
    themes = []
    for theme in mjtheme_cell:
        new_theme = theme.copy()
        if new_theme['name'] == '':
            new_theme['name'] = codemap[new_theme['code']]
        themes.append(new_theme)
    return themes
```

```python
fullnames = json_df.mjtheme_namecode.apply(fill_names)
json_df.mjtheme_namecode = fullnames
json_df.mjtheme_namecode
```
